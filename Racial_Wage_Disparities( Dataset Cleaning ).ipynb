```python
df = df[(df['INCWAGE'] > 0) & (df['WKSWORK1'] > 0) & (df['UHRSWORKT'] > 0)]
```

```python
df['EDUC'].unique()
```

```python
ambiguous = [
    'Grades 1, 2, 3, or 4',
    'Grades 5 or 6',
    'Grades 7 or 8',
    'Some college but no degree',
    'None or preschool'
]
df = df[df['EDUC'].isin(ambiguous) == False]
```

```python
df.shape
```

```python
df['EDUC'].unique()
```

```python
education_years = {
    'Grade 9': 9,
    'Grade 10': 10,
    'Grade 11': 11,

    '12th grade, no diploma': 12,
    'High school diploma or equivalent': 12,

    "Associate's degree, occupational/vocational program": 14,
    "Associate's degree, academic program": 14,

    "Bachelor's degree": 16,

    "Master's degree": 18,

    'Professional school degree': 19,
    'Doctorate degree': 21
}
df['EDUCYRS'] = df['EDUC'].map(education_years)
```

```python
df['YEAR'].value_counts().sort_index()
```

```python
df["HISPAN"].unique()
```

```python
#BINARY ENCODING FOR HISPAN
```

```python
# First replace "Not Hispanic" with 0
df['HISPAN'] = df['HISPAN'].replace({'Not Hispanic': 0})

# Now convert everything else into 1 using np.where
df['HISPAN'] = np.where(df['HISPAN'] == 0, 0, 1).astype(int)
```

```python
#BINARY ENCODING FOR CLASSWKR
```

```python
# Replace all government categories with 1
df['CLASSWKR'] = df['CLASSWKR'].replace({
    'Federal government employee': 1,
    'State government employee': 1,
    'Local government employee': 1
})

# All remaining categories become 0
df['CLASSWKR'] = np.where(df['CLASSWKR'] == 1, 1, 0).astype(int)
```

```python
# Make a panel of counts for every column by YEAR
panel_counts = df.groupby("YEAR").count().reset_index()

# Optional: save to CSV
panel_counts.to_csv("yearly_panel_counts.csv", index=False)

panel_counts
```
