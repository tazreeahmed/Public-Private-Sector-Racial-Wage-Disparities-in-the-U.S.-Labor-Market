```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cps_00008.csv")
df.head()
```

```python
df['YEAR'].value_counts().sort_index()
```

```python
df.columns
```

```python
#ENCODING
```

```python
#1. Encode for sex
```

```python
df['SEX'] = df['SEX'].replace(1, 'Male')
df['SEX'] = df['SEX'].replace(2, 'Female')
df['SEX'] = df['SEX'].replace(9, 'NIU')
```

```python
df['RACE'].unique()
```

```python
#Encode for Race
```

```python
# Step 1: Map numeric race codes to analytic groups
race_map = {
    100: 'White',
    200: 'Black',
    300: 'Native',
    650: 'Asian',
    651: 'Asian',
    652: 'Asian'
}

df["RACE_REC"] = df["RACE"].map(race_map)

# Step 2: Everything else becomes "Other"
df["RACE_REC"] = df["RACE_REC"].fillna('Other')

# Step 3: Override with Hispanic ethnicity
df["RACE_ETH"] = np.where(df["HISPAN"] == 1, 'Hispanic', df["RACE_REC"])
```

```python
df['RACE_ETH'].unique()
```

```python
#Encoding for MARST
```

```python
marst_map = {
    1: "Married, spouse present",
    2: "Married, spouse absent",
    3: "Separated",
    4: "Divorced",
    5: "Widowed",
    6: "Never married/single",
    7: "Widowed or Divorced",
    9: "NIU"
}

df['MARST'] = df['MARST'].replace(marst_map)
```

```python
#Encoding for Nativity
```

```python
nativity_map = {
    0: "Unknown",
    1: "Both parents native-born",
    2: "Father foreign, mother native",
    3: "Mother foreign, father native",
    4: "Both parents foreign",
    5: "Foreign born"
}

df['NATIVITY'] = df['NATIVITY'].replace(nativity_map)
```

```python
#Encode for region:
```

```python
region_map = {
    11: "New England Division",
    12: "Middle Atlantic Division",
    21: "East North Central Division",
    22: "West North Central Division",
    31: "South Atlantic Division",
    32: "East South Central Division",
    33: "West South Central Division",
    41: "Mountain Division",
    42: "Pacific Division",
    97: "State not identified"
}

df["REGION"] = df["REGION"].replace(region_map)
```

```python
#Encoding for UHRSWORKT
```

```python
df['UHRSWORKT'] = df['UHRSWORKT'].replace(997, "Hours vary")
df['UHRSWORKT'] = df['UHRSWORKT'].replace(999, "NIU")
```

```python
#Encoding for 
```

```python
df.isna().sum()
df = df[df["AGE"].notna()]
```

```python
#Encoding for Occupation
```

```python
occ = pd.read_csv("codes1990.csv")

occ['code'] = occ['code'].astype(int)
df['OCC1990'] = df['OCC1990'].astype(int)
df['OCC1990'] = df['OCC1990'].replace(occ.set_index('code')['label'])
```

```python
#Encoding for industry
```

```python
ind = pd.read_csv("IND1990.csv")

ind['code'] = ind['code'].astype(int)
df['IND1990'] = df['IND1990'].astype(int)
df['IND1990'] = df['IND1990'].replace(ind.set_index('code')['label'])
```

```python
#Encoding for class of worker
```

```python
classwkr = pd.read_csv("CLASSWKR.csv")

classwkr['code'] = classwkr['code'].astype(int)
df['CLASSWKR'] = df['CLASSWKR'].astype(int)
df['CLASSWKR'] = df['CLASSWKR'].replace(classwkr.set_index('code')['label'])
```

```python
#Encoding for education
```

```python
educ = pd.read_csv("EDUC.csv")

educ['code'] = educ['code'].astype(int)
df['EDUC'] = df['EDUC'].astype(int)
df['EDUC'] = df['EDUC'].replace(educ.set_index('code')['label'])
```

```python
#Encoding for Hispan
```

```python
hispan_map = {
    0:   "Not Hispanic",
    100: "Mexican",
    102: "Mexican American",
    103: "Mexicano/Mexicana",
    104: "Chicano/Chicana",
    108: "Mexican (Mexicano)",
    109: "Mexicano/Chicano",
    200: "Puerto Rican",
    300: "Cuban",
    400: "Dominican",
    500: "Salvadoran",
    600: "Other Hispanic",
    610: "Central/South American",
    611: "Central American (excluding Salvadoran)",
    612: "South American",
    901: "Do not know",
    902: "N/A"
}

df['HISPAN'] = df['HISPAN'].replace(hispan_map)
```
