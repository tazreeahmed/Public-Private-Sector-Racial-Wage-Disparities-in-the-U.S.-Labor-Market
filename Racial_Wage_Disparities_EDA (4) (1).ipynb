```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("democleaned.csv")
df.head()
```

```python
region_map = {
    11: "New England Division",
    12: "Middle Atlantic Division",
    21: "East North Central Division",
    22: "West North Central Division",
    31: "South Atlantic Division",
    32: "East South Central Division",
    33: "West South Central Division",
    41: "Mountain Division",
    42: "Pacific Division",
    97: "State not identified"
}
df["REGION_LABEL"] = df["REGION"].replace(region_map)
```

```python
df.columns
```

```python
cols_to_drop = [
    'SERIAL', 'MONTH', 'CPSID', 'ASECFLAG', 'ASECWTH',
    'EXPERIENCE_SQ', 'PERNUM', 'CPSIDP', 'CPSIDV', 'ASECWT',
    'OCC1990', 'IND1990', 'REGION_LABEL', 'LNWAGE', 'Unnamed: 0'
]

df = df.drop(cols_to_drop, axis=1, errors='ignore')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr = numeric_df.corr()

# Set plot size
plt.figure(figsize=(10, 7))

# Create heatmap with numeric values shown
sns.heatmap(
    corr,
    annot=True,        
    fmt=".2f",         
    cmap='coolwarm',   
    linewidths=0.5,
    square=False       
)

# Title and layout
plt.title("Correlation Heatmap", fontsize=18)
plt.tight_layout()

# Display the heatmap
plt.show()
```

```python
df['RACE'] = df['RACE'].astype(str).str.strip()
race_map = {
    '100': 'White',
    '200': 'Black',
    '300': 'Native',
    '650': 'Asian',
    '651': 'Asian',
    '652': 'Asian'
}
df['RACE_REC'] = df['RACE'].map(race_map).fillna('Other')
df['RACE_ETH'] = df['RACE_ETH'].where(df['RACE_ETH'] == 'Hispanic', df['RACE_REC'])

df['HOURLY_WAGE'] = df['INCWAGE'] / (df['WKSWORK1'] * df['UHRSWORKT'])

plt.figure(figsize=(10, 6))

# Bar plot: average hourly wage by race, colored by government sector
sns.barplot(
    data=df,
    x='RACE_ETH',
    y='HOURLY_WAGE',
    hue='CLASSWKR',
    estimator='mean',
    errorbar=None
)

plt.title('Average Hourly Wage by Race, Colored by Government Sector')
plt.ylabel('Average Hourly Wage')
plt.xlabel('Race')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))

# Line plot: average hourly wage by race across years
sns.lineplot(
    data=df,
    x="YEAR",
    y="HOURLY_WAGE",
    hue="RACE_ETH",
    estimator="mean",
    errorbar=None
)

plt.title("Average Hourly Wage by Race (2000–2023)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Average Hourly Wage")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Keep only Black and Native workers
df_sub = df[df['RACE_ETH'].isin(['Black', 'Native'])]

plt.figure(figsize=(10,6))
ax = sns.barplot(
    data=df_sub,
    x='HOURLY_WAGE',
    y='REGION',
    hue='RACE_ETH',
    estimator='mean',
    errorbar=None
)

plt.title('Average Hourly Wage by Region (Black vs Native)')
plt.xlabel('Average Hourly Wage')
plt.ylabel('Region')
ax.legend(title='Race')
plt.tight_layout()
plt.show()
```
